```python
%matplotlib inline

import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format

import json
import glob
import os
```

```python
data = json.load(open("/home/zeitz/experiments/bidir_astar/bidir_new.json"))
```

```python
queries = pd.DataFrame.from_records([{
    'experiment': exp['experiment'],
    'factor': exp['factor'],
    'probability': exp.get('probability', 1),
    'speed': exp.get('speed', 0),
    **algo }
    for exp in data['experiments'] for algo in exp['algo_runs']])
```

```python
queries['choose_direction_strategy'] = queries['choose_direction_strategy'].fillna('unidir')
queries['bidir_pot'] = queries['bidir_pot'].fillna('unidir')
```

```python
results = queries.groupby(['algo', 'choose_direction_strategy', 'bidir_pot', 'experiment', 'factor', 'probability', 'speed'])[['running_time_ms', 'num_queue_pops']].median().unstack([0,1,2]).dropna(axis='columns')
```

```python
table = queries.groupby(['algo', 'experiment', 'probability', 'speed', 'factor'])[['running_time_ms', 'num_queue_pops']].median().unstack(0).loc['probabilistic_scale_by_speed']
table[('running_time_ms', 'Symmetric Bidrectional A* Query')] = table[('running_time_ms', 'Symmetric Bidrectional A* Query')] / table[('running_time_ms', 'Dijkstra Query')]
table[('running_time_ms', 'Bidrectional Dijkstra Query')] = table[('running_time_ms', 'Bidrectional Dijkstra Query')] / table[('running_time_ms', 'Dijkstra Query')]
table[('num_queue_pops', 'Symmetric Bidrectional A* Query')] = table[('num_queue_pops', 'Symmetric Bidrectional A* Query')] / table[('num_queue_pops', 'Dijkstra Query')]
table[('num_queue_pops', 'Bidrectional Dijkstra Query')] = table[('num_queue_pops', 'Bidrectional Dijkstra Query')] / table[('num_queue_pops', 'Dijkstra Query')]
table
```
